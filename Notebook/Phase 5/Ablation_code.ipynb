```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Loading dataset

X_train = np.load("/Users/amarachi/Documents/AQA/Exp_3/phase3/notebook /phase3_features/X_train.npy")
X_test  = np.load("/Users/amarachi/Documents/AQA/Exp_3/phase3/notebook /phase3_features/X_test.npy")

y_train = np.load("/Users/amarachi/Documents/AQA/Exp_3/phase3/notebook /phase3_features/y_train.npy")
y_test  = np.load("/Users/amarachi/Documents/AQA/Exp_3/phase3/notebook /phase3_features/y_test.npy")

print("Loaded successfully")
print("X_train shape:", X_train.shape)
```

```python
# split features

rule_dim = 8   

# RULE FEATURES
train_rules_only = X_train[:, :rule_dim]
test_rules_only  = X_test[:, :rule_dim]

# TRANSFORMER FEATURES
train_embed_only = X_train[:, rule_dim:]
test_embed_only  = X_test[:, rule_dim:]

print("Rule shape:", train_rules_only.shape)
print("Embed shape:", train_embed_only.shape)
```

```python
# Rule-only model

rule_model = LogisticRegression(max_iter=1000)
rule_model.fit(train_rules_only, y_train)

rule_preds = rule_model.predict(test_rules_only)

rule_p = precision_score(y_test, rule_preds)
rule_r = recall_score(y_test, rule_preds)
rule_f = f1_score(y_test, rule_preds)

print("\n=== RULE-ONLY ===")
print("Precision:", round(rule_p,4))
print("Recall:", round(rule_r,4))
print("F1:", round(rule_f,4))
```

```python
#Transformer-only models


trans_model = LogisticRegression(max_iter=1000)
trans_model.fit(train_embed_only, y_train)

trans_preds = trans_model.predict(test_embed_only)

trans_p = precision_score(y_test, trans_preds)
trans_r = recall_score(y_test, trans_preds)
trans_f = f1_score(y_test, trans_preds)

print("\n=== TRANSFORMER-ONLY ===")
print("Precision:", round(trans_p,4))
print("Recall:", round(trans_r,4))
print("F1:", round(trans_f,4))
```

```python
test_probs = np.load("/Users/amarachi/Documents/AQA/Exp_3/phase4/Notebook/phase4_outputs/test_probs.npy")
```

```python
# RADAR 

radar_preds = (test_probs >= 0.5).astype(int)

radar_p = precision_score(y_test, radar_preds)
radar_r = recall_score(y_test, radar_preds)
radar_f = f1_score(y_test, radar_preds)

print("\n=== RADAR (HYBRID) ===")
print("Precision:", round(radar_p,4))
print("Recall:", round(radar_r,4))
print("F1:", round(radar_f,4))
```
